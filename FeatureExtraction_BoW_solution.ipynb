## Input Data##

Below data is another text classification, problem, ham and spam detection. A sms you recives in your mobile phone could be eighte ham or spam.

```python
texts = [
    "Pay your connection bill of 2000",
    "Win 20000 by participating in the lottery.",
    "Win 200 rupees by participating this competition at gamer.com.",
    "I will come tomorrow"
]
```

## Turn each of this sms to it's numerical vector (Bag of Words) ##

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

cv = CountVectorizer(ngram_range=(1,1))
features = cv.fit_transform(texts)
pd.DataFrame(
features.todense(),
columns = cv.get_feature_names())
```
